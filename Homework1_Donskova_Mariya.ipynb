```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def get_max_before_zero(x):
    zero = x == 0
    if len(x[1:][zero[:-1]]) == 0:
        return None
    else:
        return x[1:][zero[:-1]].max()
```

**Написать функцию replace_nan_to_means(X), принимающую матрицу X. Функция должна вернуть копию матрицы X, в которой все значения nan в каждом столбце заменены на среднее арифметическое остальных элементов столбца. В случае столбца из одних nan необходимо заменить все элементы столбца на нули. Исходная матрица X должна остаться неизменной!**  
Далее представление 3 реализации поставленной задачи
- реализация через numpy
- смешанная реализация
- совсем без numpy 

```python
import numpy as np
import random
siz = 10
A = np.random.random((siz,siz))
A[random.randint(0, siz-1)][random.randint(0, siz-1)] = np.nan


def replace_nan_to_means(X):
    mea_ = np.nanmean(X, axis=0)
    k = np.where(np.isnan(mea_))
    mea_[k] = 0
    ind = np.where(np.isnan(X))
    X[ind] = np.take(mea_, ind[1])
    return X

t = %timeit -o replace_nan_to_means(A)
```

**Написать функцию encode_rle(x), реализующую кодирование длин серий (Run-length encoding). По входному вектору x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй — сколько раз их нужно повторить.**  
Далее представление 3 реализации поставленной задачи
- реализация через numpy
- смешанная реализация
- совсем без numpy 

```python
import numpy as np
import random
from random import randint

a = np.random.randint(0, 3, 10000000)

def encode_rle(a):
    if len(a) == 1:
        return (a, np.array([1]))
    a_notexit = np.array([min(a) - 1])
    a = np.concatenate((a, a_notexit))
    mask = (a != np.roll(a, 1))
    a_bez_posl = a[mask][:-1]
    indix = np.arange(len(a))[mask]
    numb_ = np.diff(indix)
    return (a_bez_posl, numb_)

t0 = %timeit -o encode_rle(a)
```
