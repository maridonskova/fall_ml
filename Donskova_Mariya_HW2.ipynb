__10__. Сделайте столбец `trx_category_mul`, которая показывает увеличение (значение в столбце равно +1) и уменьшение средств (значение в столбце равно -1) на счете клиента банка. Операции, которые увеличивают количество средств на счете:
`C2C_IN`, `DEPOSIT`, `BACK_TRX`. Считайте, что остальные операции уменьшают количество средств на счете.

````python
def oper(trx):
    if trx == 'C2C_IN' or trx == 'DEPOSIT' or trx == 'BACK_TRX':
        trx_m = 1
    else:
        trx_m = -1
    return trx_m
df.loc[:, 'trx_category_mul'] = df.loc[:, 'trx_category'].apply(oper)
````

__11__. Приведем столбец `TRDATETIME` к типу даты.

````python
from datetime import datetime
from datetime import date

df["TRDATETIME"] = pd.to_datetime(df["TRDATETIME"], format='%d%b%y:%H:%M:%S')
df.head()
````

Извлеките из столбца `TRDATETIME` следующие признаки:
* `mday` – день (с начала месяца);
* `ymonth` – месяц;
* `year` – год;
* `wday` – день недели;
* `is_weekend` – является ли день выходным днем.

Извлеченные признаки запишите в таблицу `df_time`.

````python
df_time = pd.DataFrame(np.zeros((490513, 0)))
df_time['mday'] = df["TRDATETIME"].dt.day
df_time['year'] = df["TRDATETIME"].dt.year
df_time['ymonth'] = df["TRDATETIME"].dt.month
df_time['wday'] = df['TRDATETIME'].dt.weekday

def func_(wday):
    if wday < 5:
        return False
    else:
        return True
df_time.loc[:, 'is_weekend'] = df_time.loc[:,'wday'].apply(func_)


df_time.head()
````
